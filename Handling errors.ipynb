```python
#We handle this error using try..except statement:
try:   #The code inside the try block is first run
  x = int(input("Enter a number: "))
except:  #For error's exception
  print("This\'s not a valid number")
```

When we use (try..except) and then enter a not valid input, we will not get error. The program will be run and the code block in except will be executed. If there is a code left, this will be executed as well.
