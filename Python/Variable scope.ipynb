**Variable scope** refers to which parts of a program a variable can be referenced, or used, from.

It's important to consider scope when using variables in functions. If a variable is created inside a function, it can only be used within that function. Accessing it outside that function is not possible.

````python
def some_function():
  word = "Hello"
  
print(word)
````

Ih the previous example, we got that error (NameError: name 'word' is not defined) as word is a local variable. so we can't use it outside the function---> word variable scope is local---->This means you can use the same name for different variables that are used in different functions.

````python
def some_function():
    word = "hello"

def another_function():
    word = "goodbye"
````

````python
# This works fine---> (word) variable is a global scope
word = "hello"

def some_function():
    print(word)

some_function()
````

Python doesn't allow functions to modify variables that are outside the function's scope. A better way would be to pass the variable as an argument and reassign it outside the function

Python doesn't allow functions to modify variables that aren't in the function's scope.

````python
egg_count = 0

def buy_eggs():
    egg_count += 12 # purchase a dozen eggs

buy_eggs()
#What is the result of running this code? If you aren't sure, try running it on your own computer!
#answer ----> This causes an UnboundLocalError
````

````python
egg_count = 0

def buy_eggs(count):
    return count + 12  # purchase a dozen eggs

egg_count = buy_eggs(egg_count)
print(egg_count)
````
