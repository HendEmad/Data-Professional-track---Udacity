

*   Documentation strings (docstring) ----> A type of comment used to explain the purpose of a function and how it should be used.
*   It is a part of good code practice as it make the code more readable and assist you and future users to uncerstand the code you produced.
*    Example: 

```
def population_density(population, land_area):

#explanation of function(one-line summary)
    """Calculate the population density of an area.    

#explanation of the function's arguments, stating the purpose and types each one
    INPUT:
    population: int. The population of that area
    land_area: int or float. This function is unit-agnostic, if you pass in values in terms
    of square km or square miles the function will return a density in those units.

 #description of the function's output.
    OUTPUT: 
    population_density: population / land_area. The population density of a particular area.
    """
    return population / land_area
```
* note: Docstrings are surrounded by triple quotes. The first line of the docstring is a brief explanation of the function's purpose. If you feel that this is sufficient documentation you can end the docstring at this point; single line docstrings are perfectly acceptable




## Quiz: Write a Docstring
Write a docstring for the readable_timedelta function you defined earlier! Remember the way you write your docstrings is pretty flexible!
```
def readable_timedelta(days):
    # insert your docstring here

    weeks = days // 7
    remainder = days % 7
    return "{} week(s) and {} day(s)".format(weeks, remainder)
```



````python
#answer:
def readable_timedelta(days):
    """
    Return a string of the number of weeks and days included in days.

    Parameters:
    days -- number of days to convert (int)

    Returns:
    string of the number of weeks and days included in days
    """
    weeks = days // 7
    remainder = days % 7
    return "{} week(s) and {} day(s)".format(weeks, remainder)
````
